## График функции


```python
from sympy import *
from sympy.plotting import plot
from matplotlib import *
init_printing()
x = Symbol('x')
f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
plot(f,(x,-100,100),title = 'график функции')
```

## корни уравнения


```python
solve(Eq((f,0),(x,-10,10))
```

```python
g = f.diff(x)
plot(g,(x,-10,10),title = 'график функции')
```

```python
print(g)
```

## вершина функции

```python
solve(Eq(g,0),x)
```

## минимальное значение функции

```python
f.subs(x,-1)
```
